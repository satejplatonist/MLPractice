```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r'D:\MLPractice\Datasets\Titanic.csv')
```

```python
data.head()
```

```python
#to display the total number columns present in the dataset
pd.set_option('display.max_columns', None)
```

```python
data.isnull().sum()
```

```python
data.isnull().mean()
```

```python
data.isnull().mean().plot.bar(figsize=(8,8))
plt.ylabel('Percentage of missing values')
plt.xlabel('variables')
plt.title('Quantifying missing values')
```

```python
data.isnull().sum().plot.bar(figsize=(8,8))
plt.xlabel('variables')
plt.ylabel('Total missing values per col')
plt.title('Quantifying missing values')
```
