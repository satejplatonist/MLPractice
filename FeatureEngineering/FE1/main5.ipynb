```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.datasets import fetch_california_housing
```

```python
caldataset = fetch_california_housing()
```

```python
dataset = pd.DataFrame(caldataset.data)
```

```python
dataset.columns = [caldataset.feature_names]
```

```python
dataset.head()
```

```python
print(caldataset.DESCR)
```

```python
np.random.seed(29)
x = np.random.rand(200)
data = pd.DataFrame([x]).T
data.columns = ['x']
data.head()
```

```python
sns.distplot(dataset['HouseAge'], bins=30)
```

```python
sns.distplot(dataset['Latitude'], bins=30)
```

```python
stats.probplot(data['x'], dist="norm", plot=plt)
plt.show()
```
